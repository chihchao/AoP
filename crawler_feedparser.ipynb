```python
import feedparser

feed_url = "https://tw.news.yahoo.com/rss"  # Replace with the actual feed URL
feed_data = feedparser.parse(feed_url)
for entry in feed_data.entries[:5]:
    print(entry.title)
    print(entry.summary)
    print(entry.link)
    print("---")
```

```python
import pandas as pd
df = pd.DataFrame(feed_data.entries)
df = df[['title','summary','link']]
df.head(5)
```

```python
df.to_csv('yahoo_news.csv', index=False, encoding='utf-8-sig')
```
